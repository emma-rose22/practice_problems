```
def change_emojis(x):
  #avoiding nans
  if type(x) != float:
    #iterating through each letter in the cell
    for letter in x:
      #this section is creating a window of three characters to search through the cell
      #most of the emojis seem to be 3 characters long, this can be improved upon
      #to catch longer, less frequent emojis
      one_window = x.index(letter) - 1
      three_window = x.index(letter) + 1
      try:
        #if not at the beginning or end, this sets up our window
        window = [x[one_window] + letter + x[three_window]]
      except:
        #if it is at the beginning or end, we need to simply leave that out
        #of the window
        if one_window == -1:
          window = [letter + x[three_window]]
        if three_window == len(x) + 1:
          window = [x[one_window] + letter]

      #this section is where we apply the window and switch out the characters with the emoji

      #all of the emojis should have a weird O and Y in front, but many are missing this
      #the first section tries to put in emoji, assuming the O and Y are present
      #the second adds them in
      try:
        value = str(window[0])
        try:
          #getting rid of extra weird commas
            value = value.strip('¸')
        except:
          pass
        #stripping away whitespace 
        value = value.strip()
        if value in dic1.keys():
          #this assumes the value is the complete emoji code
          index = x.index(value)
          x = x[:index] + value.encode('cp1252').decode('utf8') + x[index+1:]
          print(x)

        #this adds in the O and Y to the window
        if window[0] not in dic1:
          window = ['ðŸ'] + window
          window = [window[0] + window[1]]
          value = str(window[0])
          #again, cleaning the value
          try:
            value = value.strip('¸')
          except:
            pass
          value = value.strip()

          #checking in the dictionary for widened window, switching out if present
          if value in dic1.keys():
            index = x.index(value[2:])
            x = x[:index] + value.encode('cp1252').decode('utf8') + x[index+2:]
            print(x)
      except:
        #if neither section could find the emoji
        print(f'ERROR: Could not encode/decode this emoji {value}')
  return x

  #the best thing that could be done to improve the performance of this is to improve the dictionary
  #the values being missed are either not there, or there incorrectly

  #alternatively, you could keep track of the emojis the dictionary is rejecting, but should actually work
  #and switch out all the ones that don't work in the column with the same exact one that does work
  #then run the function again
  #and you may be able to improve the amount of emojis switched

  #so far all the ones I am seeing that don't work, are 'in' the dict but won't actually switch over
  #which is the phenomena we observed earlier :/

  #of course, I haven't had time to fully evaluate it's performance, as I said above the window
  #could likely be improved.

  #anyways, good luck with this stuff :) I am glad I could help.
```
